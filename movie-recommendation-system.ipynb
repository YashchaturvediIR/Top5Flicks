```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):  #ast.literal_eval will convert obj string into list form
        L.append(i['name'])
        return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast

def convert2(obj):   # for get cast name up to 3 dict (Main Heros) as after 3 that is of no use
    L = []
    counter = 0
    for i in ast.literal_eval(obj):  # ast.literal_eval will convert obj string into list form
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
def fetch_director(obj):
    # Similar to convert function but for fetching director names
    L = []
    for i in ast.literal_eval(obj):  # ast.literal_eval will convert obj string into list form
        if i['job'] == 'Director':  #job is dict parameter 
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())  #To make string into a list format just like other lists above 
```

```python
# Apply the lambda function to remove spaces in the strings within the lists
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
    
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```
